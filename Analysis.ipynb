```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('region_traffic.csv')
```

```python
plt.figure(figsize=(10, 6))
plt.hist(data['all_motor_vehicles'], bins=20, color='blue', alpha=0.7)
plt.title('Distribution of Total Motor Vehicles')
plt.xlabel('Total Motor Vehicles')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['region_name'], data['all_motor_vehicles'], color='orange')
plt.title('Total Motor Vehicles by Region')
plt.xlabel('Region')
plt.ylabel('Total Motor Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data['road_category_name'], data['all_motor_vehicles'], color='red')
plt.title('Total Motor Vehicles by Road Category')
plt.xlabel('Road Category')
plt.ylabel('Total Motor Vehicles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
grouped_data = data.groupby('region_name').sum()[['pedal_cycles', 'cars_and_taxis', 'buses_and_coaches', 'lgvs', 'all_hgvs']]

plt.figure(figsize=(12, 8))
grouped_data.plot(kind='bar', stacked=True, colormap='viridis')
plt.title('Variance of Different Vehicle Types Across Regions')
plt.xlabel('Region')
plt.ylabel('Total Count')
plt.xticks(rotation=45)
plt.legend(title='Vehicle Type')
plt.tight_layout()
plt.show()
```

```python
data = pd.read_csv('traffic.csv')
data.head()
```

```python
data['Date'] = pd.to_datetime(data['DateTime']).dt.date
data['Hour'] = pd.to_datetime(data['DateTime']).dt.hour
data['Minute'] = pd.to_datetime(data['DateTime']).dt.minute
data['Second'] = pd.to_datetime(data['DateTime']).dt.second
```

```python
hourly_congestion = data.groupby('Hour')['Vehicles'].mean()
plt.figure(figsize=(10, 6))
hourly_congestion.plot(kind='bar', color='skyblue')
plt.title('Hourly Congestion Levels')
plt.xlabel('Hour of the Day')
plt.ylabel('Average Vehicles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
date_traffic = data.groupby('Date')['Vehicles'].sum()
plt.figure(figsize=(10, 6))
date_traffic.plot(kind='line', marker='o', color='green')
plt.title('Date-wise Traffic')
plt.xlabel('Date')
plt.ylabel('Total Vehicles')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
junction_traffic = data.groupby('Junction')['Vehicles'].sum()
plt.figure(figsize=(8, 5))
junction_traffic.plot(kind='bar', color='orange')
plt.title('Junction-wise Traffic')
plt.xlabel('Junction')
plt.ylabel('Total Vehicles')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```

```python
data['Month'] = pd.to_datetime(data['DateTime']).dt.month
monthly_traffic = data.groupby('Month')['Vehicles'].sum()
plt.figure(figsize=(10, 6))
monthly_traffic.plot(kind='bar', color='orange')
plt.title('Monthly Traffic Volume')
plt.xlabel('Month')
plt.ylabel('Total Vehicles')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.show()
```
